```python
import matplotlib.pyplot as plt
```

```python
p1_x = [0,-2,-2,-2,2]
p1_y = [2,2,2,-2,-2]
p2_x = [0,2,2]
p2_y = [-2,-2,0]
x = [1,-2]
y = [3,-3]
```

```python
plt.plot(p1_x,p1_y, linestyle="none", marker="*", color="red", label="p1")
plt.plot(p2_x,p2_y, linestyle="none", marker="o", color="black", label="p2")
plt.plot(x,y, marker="p")
plt.title("Grafica Punto 8")
#plt.legend(loc=3)
plt.grid()
plt.show()
```
