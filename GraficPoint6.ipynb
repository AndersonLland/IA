```python
import matplotlib.pyplot as plt
```

```python
p1_x = [-1,0,1]
p1_y = [1,0,-1]
p2_x = [1,0]
p2_y = [-0,1]
x = [-1.5,2]
y = [2,-1.5]
```

```python
plt.plot(p1_x,p1_y, linestyle="none", marker="*", color="red", label="p1")
plt.plot(p2_x,p2_y, linestyle="none", marker="o", color="black", label="p2")
plt.plot(x,y, marker="p")
plt.title("Grafica Punto 6")
#plt.legend(loc=3)
plt.grid()
plt.show()
```
