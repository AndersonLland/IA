```python
import matplotlib.pyplot as plt
```

```python
p1_x = [0.5,-1,-0.5]
p1_y = [1,-1,-1]
p2_x = [-0.5,1,0.5]
p2_y = [-1,1,-1]
x = [0.75,-0.75]
y = [1,-1]
```

```python
plt.plot(p1_x,p1_y, linestyle="none", marker="*", color="black", label="Naranjas")
plt.plot(p2_x,p2_y, linestyle="none", marker="o", color="red", label="Manzanas")
plt.plot(x,y, marker="p")
plt.title("Grafica Punto 10")
plt.legend(loc=5)
plt.grid()
plt.show()
```
