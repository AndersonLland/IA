```python
import matplotlib.pyplot as plt
```

```python
p1_x = [1,1]
p1_y = [1,2]
p2_x = [2,2]
p2_y = [-1,0]
x1 = [-1,1]
y1 = [3,-3]
p1_x1 = [-1,-2]
p1_y1 = [2,1]
p2_x2 = [-1,-2]
p2_y2 = [-1,-2]
x2 = [3,-3]
y2 = [1,-1]
```

```python
plt.plot(p1_x,p1_y, linestyle="none", marker="*", color="red", label="p1")
plt.plot(p2_x,p2_y, linestyle="none", marker="*", color="black", label="p2")
plt.plot(x1,y1, marker="p")
plt.title("Grafica Punto 7")
plt.plot(p1_x1,p1_y1, linestyle="none", marker="o", color="blue", label="p1")
plt.plot(p2_x2,p2_y2, linestyle="none", marker="o", color="yellow", label="p2")
plt.plot(x1,y1, marker="p")
plt.plot(x2,y2, marker="p")
plt.grid()
plt.show()
```
